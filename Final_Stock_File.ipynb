```python
import pandas as pd
from fuzzywuzzy import process

# Read the CSV file into DataFrame
stock_df = pd.read_csv(r"C:\Users\DELL\Desktop\Projects\LOC 6.0\Categories_Mapping.csv")
sector_definitions = pd.read_csv(r"C:\Users\DELL\Desktop\Projects\LOC 6.0\Sector_Definitions.csv")
```

```python
sector_definitions.head(5)
```

```python
# Function to find the most similar sector or industry
def find_similar_sector_or_industry(query, choices):
    match = process.extractOne(query, choices)
    return match[0] if match else None

# Add definitions to stock_df
for index, row in stock_df.iterrows():
    sector = row['Sector']
    industry = row['Industry']
    
    # Find similar sector in sector_definitions
    similar_sector = find_similar_sector_or_industry(sector, sector_definitions['Sector'])
    
    # Find similar industry in sector_definitions
    similar_industry = find_similar_sector_or_industry(industry, sector_definitions['Industry'])
    
    # Use similar sector or industry to get definition
    if similar_sector:
        definition = sector_definitions.loc[sector_definitions['Sector'] == similar_sector, 'Definition'].iloc[0]
    elif similar_industry:
        definition = sector_definitions.loc[sector_definitions['Industry'] == similar_industry, 'Definition'].iloc[0]
    else:
        definition = 'No definition found'
    
    # Assign definition to stock_df
    stock_df.at[index, 'Definition'] = definition
```
